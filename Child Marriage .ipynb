# **PERCENTAGE OF MARRIAGE CATEGORIES BY CONTINENT**

```python
Omo
```

# **GLOBAL PERCENTAGE FOR CATEGORY 15**

```python
# to select columns for category 15
Global_15_Percentage = Omo [["Continent/region", "Married by 15 Female"]]
Global_15_Percentage
```

```python
continent_totals = Global_15_Percentage.groupby('Continent/region')['Married by 15 Female'].sum()
continent_totals
```

```python
# To calculate the percentage of each continent under the Married by 15 Category
continent_percentage = (continent_totals / continent_totals.sum() * 100).round().astype(int)
continent_percentage
```

```python
# To convert the output into a DataFrame
continent_percentage_Global_15_Percentage = pd.DataFrame({'Percentage of Category 15': continent_percentage})
continent_percentage_Global_15_Percentage
```

```python
# To import the required packages for charts, color intensity.
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
```

```python
values =continent_percentage_Global_15_Percentage['Percentage of Category 15'].round().astype(int)
```

```python
fig = go.Figure(go.Pie(
    labels=continent_percentage_Global_15_Percentage.index,
    values=values, marker=dict(colors= px.colors.sequential.Blues_r),
))
```

```python
fig.update_layout(
    title='Percentage of Category 15 by continent', width=800,
    height=600
)
```

# **Percentage Distribution of African Countries contributing to Child Marriage**

```python
# To select the Required columns for Analysis
Omo= child_marriage[['countries and area', 'Married by 15 Female', 'Married by 18 Female','Continent/region']]
Omo
```

```python
Africa_Omo = Omo[Omo['Continent/region'] == 'Africa']
Africa_Omo
```

```python
Africa_15 = Africa_Omo[['countries and area', 'Married by 15 Female', 'Married by 18 Female']]
Africa_15
```

```python
Africa_15 = Africa_Omo[['countries and area', 'Married by 15 Female']]
Africa_15
```

```python
Africa_15_indexed = Africa_15.set_index('countries and area', inplace=True)
```

```python
fifteen = 'Married by 15 Female'
```

```python
Sum_15 = Africa_15[fifteen].sum()
Sum_15
```

```python
Africa_15['percentage'] = round((Africa_15[fifteen] / Sum_15) * 100).astype(int)
Africa_15['percentage']
```

```python
Africa_15_indexed.reset_index(inplace=True)
Africa_15_indexed
```

```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import plot
```

```python
blue_colorscale = [(0, '#F0F8FF'), (1, '#00008B')]  # Light to dark blue
```

```python
# Create a choropleth trace
fig = go.Figure(data=go.Choropleth(
    locations=Africa_15_indexed['countries and area'],
    z=Africa_15_indexed['percentage'],
    locationmode='country names',
    colorscale=blue_colorscale,
    colorbar_title='Percentage'
))
```

```python
fig.update_layout(
    title_text='Percentage distribution of countries for Category 15 in Africa',
    geo=dict(
        showcoastlines=True,
        projection_type='equirectangular'
    )
)
```

```python
fig.update_layout(width=1100, height=600)
```
